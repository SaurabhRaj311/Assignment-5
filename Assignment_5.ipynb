#Question 1: What is a REST API?
Explain difference between:
● GET
● POST
● PUT
● DELETE
Why is FastAPI preferred over Flask for modern APIs?

**Answers-**

A REST API (Representational State Transfer Application Programming Interface) is a way for different software systems to communicate over the internet using standard HTTP methods. It allows the clients (like web apps or mobile apps) to interact with the  servers and  to perform operations such as fetching, creating, updating, or deleting of the data.
REST APIs are widely used because they are simple, scalable, and stateless, meaning each request is independent and contains all of the necessary information.

**HTTP Methods in REST API-**

**1) GET**
GET is used to retrieve data from the server.
It does not modify data.
It is safe and idempotent
Example:
Fetching user details from a database.

**2) POST**
POST is basically used to create new data on the server.
Sends data in request body.
Used for submitting forms or creating new records.
Example:
Creating a new user account

**3) PUT**
PUT is used to update existing data completely.
It Replaces the entire resource
Idempotent (same request repeated = same result)
Example:
Updating the full user profile

**4. DELETE**
DELETE is used to remove data from the server.
It Permanently deletes a resource.
Example:
Deleting a user account

**Why FastAPI is Preferred Over Flask for Modern APIs?**

FastAPI is considered better than Flask for modern API development because it offers higher performance, automatic validation, and also the better developer experience.

Following are the Key Advantages:

**1) High Performance**
FastAPI is built on ASGI and Starlette, which makes it faster than Flask in handling asynchronous requests.

**2) Automatic Data Validation**
FastAPI uses Pydantic models to validate request data automatically, reducing manual error handling.

**3) Type Hint Support**
FastAPI uses Python type hints, which improves code readability and reduces the bugs.

**4) Automatic API Documentation**
FastAPI automatically generates interactive API documentation (Swagger UI and ReDoc), which is very helpful for the testing and also for the development.

**5) Asynchronous Support**
FastAPI supports async/await natively, making it suitable for the high-performance applications.


**Conclusion-**
REST APIs enables the communication between systems using different standard HTTP methods like GET, POST, PUT, and DELETE. And Among modern frameworks, FastAPI is preferred over Flask because of its speed, built-in validation, async support, and automatic documentation, making it highly suitable for scalable and production-grade APIs.

```python
from fastapi import FastAPI

app = FastAPI()

courses = {
    1: {"name": "Python", "level": "Beginner"},
    2: {"name": "Data Science", "level": "Intermediate"},
    3: {"name": "Machine Learning", "level": "Advanced"}
}

@app.get("/")
def home():
    return {"message": "Welcome to Course API"}

@app.get("/courses")
def get_courses(level: str = None):
    if level:
        return {cid: c for cid, c in courses.items() if c["level"] == level}
    return courses

@app.get("/courses/{course_id}")
def get_course(course_id: int):
    return courses.get(course_id, {"error": "Course not found"})
```

```python
#FastAPI code with validation
from fastapi import FastAPI, Path
from typing import Literal, Optional

app = FastAPI()

courses = {
    1: {"name": "Python", "level": "beginner"},
    2: {"name": "Data Science", "level": "intermediate"},
    3: {"name": "Machine Learning", "level": "advanced"}
}

@app.get("/courses/{course_id}")
def get_course(course_id: int = Path(..., gt=0)):
    return courses.get(course_id, {"error": "Course not found"})

@app.get("/courses")
def get_courses(
    level: Optional[Literal["beginner", "intermediate", "advanced"]] = None
):
    if level:
        return {k: v for k, v in courses.items() if v["level"] == level}
    return courses
```

**Explanation**

FastAPI provides built-in validation using Path and Literal types. It ensures that only valid data is accepted by the API.
```python
#creating pydantic model
from pydantic import BaseModel, Field

class Course(BaseModel):
    title: str
    price: float = Field(..., ge=0)   # minimum price = 0
    duration: int = Field(..., gt=0)   # duration must be > 0
    published: bool = False            # default value = False
```

```python
#invalid output
Course(
    title="ML Basics",
    price=-100,
    duration=10
)
```

```python
#another invalid case
Course(
    title="Deep Learning",
    price=1000,
    duration=0
)
```

3) reduces manual error checking

**How JWT works:**
a) User logs in
b) Server verifies credentials
c) Server generates JWT token
d) Token is sent to user
e) User sends token in every request
f) Server verifies token instead of storing session

 **Structure of JWT:**
**JWT has 3 parts:**

Header.Payload.Signature

**Key Features**
- secure
- compact
- self contained
- works over HTTP easily

**3) Why staeless authentication is scalable?**
Stateless means server does not store user session data and instead all information is inside the token (JWT)

**Reason for it's scalability-**

**a) No server memory usage**
Server does not store session data → saves resources

 **b) Easy horizontal scaling**
Multiple servers can verify token without sharing session data

**c) Faster performance**
No database lookup required for session validation

 **d) Works well in distributed systems**
Microservices can independently verify JWT

**e) Reduces server load**
Less dependency on central session storage

```python
#sample fastapi Test case using TestClient
from fastapi import FastAPI
from fastapi.testclient import TestClient

app = FastAPI()

@app.get("/hello")
def hello():
    return {"message": "Hello World"}

# Create test client
client = TestClient(app)

# Test case
def test_hello():
    response = client.get("/hello")
    assert response.status_code == 200
    assert response.json() == {"message": "Hello World"}

# Run test manually
test_hello()
print("Test Passed Successfully")
```

```python
#1) example basemodel
from pydantic import BaseModel

class Course(BaseModel):
    title: str
    price: float

course = Course(title="Python", price=499.99)
print(course)
```

```python
#2) Field example
from pydantic import BaseModel, Field

class Course(BaseModel):
    title: str
    price: float = Field(..., ge=0)


course = Course(title="Python", price=500)
print(course)
```

```python
#3) validator example
from pydantic import BaseModel, validator

class Course(BaseModel):
    title: str

    @validator("title")
    def title_must_not_be_empty(cls, value):
        if value.strip() == "":
            raise ValueError("Title cannot be empty")
        return value

        course = Course(title="Data Science")
print(course)
```

```python
#4) root_validator example (updated to model_validator for Pydantic V2)
from pydantic import BaseModel, Field, ValidationError, model_validator

class Course(BaseModel):
    price: float = Field(..., ge=0)
    discount: float = Field(..., ge=0)

    @model_validator(mode='after')
    def check_discount(self):
        if self.discount > self.price:
            raise ValueError("Discount cannot exceed price")
        return self

# Valid case
valid_course = Course(price=1000, discount=200)
print(f"Valid Course: {valid_course}")

# Invalid case: Discount exceeds price
try:
    invalid_course = Course(price=100, discount=200)
except ValidationError as e:
    print(f"Invalid Course (Expected Error): {e}")
```

#Question 10:Use Pydantic BaseSettings to:
● Store database URL
● Store secret key
● Store debug flag

Why should secrets never be hardcoded?

**Answers-**

```python
#1) import BaseSetting
# Install pydantic-settings package, as BaseSettings has moved.
!pip install pydantic-settings

# Import BaseSettings from the new pydantic_settings package.
from pydantic_settings import BaseSettings
```

```python
#2) create setting class
from pydantic_settings import BaseSettings, SettingsConfigDict

class Settings(BaseSettings):
    database_url: str
    secret_key: str
    debug: bool = False

    model_config = SettingsConfigDict(env_file=".env")
```

```python
#3) create .env file
import os

# Create a .env file with the specified configurations
with open(".env", "w") as f:
    f.write("DATABASE_URL=postgresql://user:password@localhost/mydb\n")
    f.write("SECRET_KEY=my_super_secret_key\n")
    f.write("DEBUG=True\n")

# Instantiate the Settings model, which will load from .env
settings = Settings()

print(f"Database URL: {settings.database_url}")
print(f"Secret Key: {settings.secret_key}")
print(f"Debug Mode: {settings.debug}")

# Clean up the .env file after demonstrating (optional for Colab)
os.remove(".env")
```

```python
# 4) load the settings
import os

# Re-create the .env file for this execution context
with open(".env", "w") as f:
    f.write("DATABASE_URL=postgresql://user:password@localhost/mydb\n")
    f.write("SECRET_KEY=my_super_secret_key\n")
    f.write("DEBUG=True\n")

settings = Settings()

print("Database URL:", settings.database_url)
print("Secret Key:", settings.secret_key)
print("Debug Mode:", settings.debug)

# Clean up the .env file after demonstrating (optional for Colab)
os.remove(".env")
```

**Why should secrets never be hardcoded?**

**1) Security Risk**
- If secrets are written directly in code, anyone with access to the source code can view them.

 **2) Accidental Exposure**
- Hardcoded secrets may be pushed to public repositories such as GitHub.

 **3) Difficult to Maintain**
- Changing secrets requires modifying the code and redeploying the application.

 **4) Different Environments Need Different Values**
- Development, testing, and production environments often use different credentials.

 **5) Follows Best Practices**
- Using environment variables keeps sensitive information separate from application logic

**Conclusion-**
BaseSettings provides a secure and organized way to manage application configuration. It allows sensitive information such as database URLs and secret keys to be stored outside the codebase. Secrets should never be hardcoded because doing so increases security risks, reduces flexibility, and violates industry best practices.
